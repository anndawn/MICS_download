```python
# walk through the downloaded files folder, put joined data into a dataframe for each survey
def walk_4_data(datapath):
#   all_obj used to store all the dataframes for all surveys
    all_obj={}
    for root, dirs, files in os.walk(datapath, topdown=False):
        for dire in dirs:
            if (dire =='microdata') &('MICS5' in os.path.join(root, dire)):
                current_dir=os.path.join(root, dire)
#               walk through the SAV files in microdata folder
                for name in os.listdir(current_dir): 
#                       SPSS file path
                    path=os.path.join(current_dir, name)
#                       grandparent folder of the file path, survey_path contains the survey name
                    survey_path=Path(path).parents[1]
#                       Basename is the survey name
                    survey=os.path.basename(survey_path)
                    if name.lower().endswith('hl.sav'):
                        df_hl, metadata_hl = pyreadstat.read_sav(path, apply_value_formats=True,usecols=['HH1','HH2','HL1','HL7','ED3','ED4A'])
                        df_hl.rename(columns={'HL1':'LN'},inplace=True)
                    elif name.lower().endswith('wm.sav'):
#                       Get data from SPSS file and add to dfs
                        df_sex, metadata_sex = pyreadstat.read_sav(path, apply_value_formats=True,usecols=['HH1','HH2','LN','WB3','WB4'])

#                       merge different module lieracy information into one dataframe
                df_final = pd.merge(df_hl,df_sex,how='right',on=['HH1','HH2','LN'])
                df_final2 = df_final[~df_final['WB4'].isna()]
                ed_cats=df_final2['ED4A'].cat.categories
                wb_cats=df_final2['WB4'].cat.categories
                allcats=[]
                allcats.extend(ed_cats)
                allcats.extend(wb_cats)
                for cat in allcats:
                    if cat not in ed_cats:
                        df_final2['ED4A'].cat.add_categories(cat,inplace=True)
                    if cat not in wb_cats:
                        df_final2['WB4'].cat.add_categories(cat,inplace=True)


                if 'ED4A' in df_final2.columns:
                    records=df_final2[df_final2['ED4A']!=df_final2['WB4']].shape[0]
                    if records!=0:
                        print(df_final2[df_final2['ED4A']!=df_final2['WB4']],metadata_hl.column_names_to_labels,metadata_sex.column_names_to_labels)
                else:
                     print('noED3'+survey)
```
